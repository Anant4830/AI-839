```python
# Function to extract numerical values from status columns, similar to checking and savings
def extract_min_value(status):
    # Extracting the lowest number from the string, handling cases like 'no checking'
    num = pd.to_numeric(status.str.extract(r'(\d+)')[0], errors='coerce')
    num.fillna(0, inplace=True)  # Replace missing values with 0 for 'no checking'/'no known savings'
    return num
```

```python
# Apply the function to 'checking_status', 'savings_status', and 'employment'
dataset['checking_min'] = extract_min_value(dataset['checking_status'])
dataset['savings_min'] = extract_min_value(dataset['savings_status'])
dataset['employment_min'] = extract_min_value(dataset['employment'])
```

```python
# Create bins for checking_status, savings_status, and employment
checking_bins = [0, 200, 500, 1000, float('inf')]
savings_bins = [0, 100, 500, 1000, float('inf')]
employment_bins = [0, 1, 4, 7, float('inf')]
```

```python
# Adjust labels to match the number of bins - 1
checking_labels = ['No Checking', '0-200', '200-500', '500-1000', '>1000']
savings_labels = ['No Savings', '0-100', '100-500', '500-1000', '>1000']
#employment_labels = ['Unemployed', '1<=X<4', '4<=X<7', '>=7']
employment_labels = ['Unemployed', '1-4', '4-7', '>7']
```

```python
# Bin the values using pd.cut()
dataset['checking_status_binned'] = pd.cut(dataset['checking_min'], bins=checking_bins, labels=checking_labels[:-1], right=False)
dataset['savings_status_binned'] = pd.cut(dataset['savings_min'], bins=savings_bins, labels=savings_labels[:-1], right=False)
dataset['employment_binned'] = pd.cut(dataset['employment_min'], bins=employment_bins, labels=employment_labels, right=False)
```

```python
# Check the results
print(dataset[['employment_min', 'employment_binned', 'employment']].head())
```

```python
# Analyze how many entries fall into each bin
checking_status_counts = dataset['checking_status_binned'].value_counts()
savings_status_counts = dataset['savings_status_binned'].value_counts()
employment_counts = dataset['employment_binned'].value_counts()
```
